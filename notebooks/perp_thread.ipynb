```python
perp_pool_data = perp.pool_data.copy()
perp_pool_data
```

```python
# add reward data
perp_pool_data['weekly_reward_tokens_old'] = 5000

perp_pool_data['weekly_reward_tokens_new'] = 3000
perp_pool_data.loc[['vAAVE','vBNB','vBTC','vFLOW','vPERP'],'weekly_reward_tokens_new'] = 5000
perp_pool_data.loc['vETH','weekly_reward_tokens_new'] = 27000
```

```python
# price data
op_price = 1.16
perp_price = 0.70

weekly_total = perp_pool_data.weekly_reward_tokens_old.sum()
print(f'Total Emissions: ${weekly_total*(op_price+perp_price)}')
print(f'OP Emissions: ${round(weekly_total*(op_price),0)}')
print(f'PERP Emissions: ${round(weekly_total*(perp_price),0)}')
```

```python
# crude reward APRs
perp_pool_data['reward_apr_old'] = (perp_pool_data.weekly_reward_tokens_old * (op_price+perp_price) / perp_pool_data.pool_tvl / (7/365)).round(3)
perp_pool_data['reward_apr_new'] = (perp_pool_data.weekly_reward_tokens_new * (op_price+perp_price) / perp_pool_data.pool_tvl / (7/365)).round(3)

perp_pool_data[['pool_tvl','weekly_reward_tokens_old','weekly_reward_tokens_new','reward_apr_old','reward_apr_new']]
```
